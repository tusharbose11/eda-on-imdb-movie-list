# Data Cleaning (Removing The Unused Information From The Dataset)

```python
refined_df = df.drop(['budget_adj','revenue_adj','overview','imdb_id','homepage','tagline','popularity','director','keywords','release_date','vote_count','vote_average','cast','genres'],axis =1)
```

```python
refined_df
```

```python
#Removing null values
refined_df.dropna(inplace=True)
refined_df
```

```python
#A seperate dataframe for questions related to budget
refined_df_budget = refined_df[refined_df['budget']>0]
print(refined_df_budget.shape)
```

```python
#A seperate dataframe for questions related to revenue
refined_df_revenue = refined_df[refined_df['revenue']>0]
print(refined_df_revenue.shape)
```

```python
#A seperate dataframe for questions related to production company
prod_company = refined_df_revenue.drop(['budget','runtime','release_year'],axis =1)
prod_company
```

Since now we have the columns, rows and format of the dataset in right way, its time to investigate the data for the questions asked.

# Exploratory Data Analysis

# Question 1 : Which are the movies with most and least earned revenue?

```python
def find_minmax(x):
    #use the function 'idmin' to find the index of lowest revenue movie.
    min_index = refined_df_revenue[x].idxmin()
    #use the function 'idmax' to find the index of Highest revenue movie.
    high_index = refined_df_revenue[x].idxmax()
    high = pd.DataFrame(refined_df_revenue.loc[high_index,:])
    low = pd.DataFrame(refined_df_revenue.loc[min_index,:])
    #print the movie with high and low revenue
    print("Movie Which Has Highest "+ x + " : ",refined_df_revenue['original_title'][high_index])
    print("Movie Which Has Lowest "+ x + "  : ",refined_df_revenue['original_title'][min_index])
    return pd.concat([high,low],axis = 1)
find_minmax('revenue')
```

# Question 2 : What is the average runtime of movies in the year 2006?

```python
refined_df.groupby('release_year')
refined_df_filtered = refined_df.query('release_year == 2006')
mean = refined_df_filtered.groupby('release_year').mean()['runtime']
mean
```

# Question 3 : Name any 3 production companies which have invested money in worse revenue movies?

```python
tester = pd.DataFrame(prod_company.groupby("production_companies").mean()['revenue'])
```

```python
tester
```

```python
#Assumption: Movies w least revenue made are taken as context in order to find the production companies.
worst_reader = prod_company.nsmallest(3, ['revenue'])
Top3_worst_prod_comp = worst_reader['production_companies']
Top3_worst_prod_comp.to_frame()
```

```python
print("Top 3 worst production companies : ",worst_reader['production_companies'])
```

# Question 4 : Which are the movies with the third lowest and third highest budget?

```python
low_reader = refined_df_budget.nsmallest(8, ['budget'])
low_reader
```

```python
#Assumption: Since, 3 movies are tied for the 3rd least budget that is why there are 3 answers.
third_lowest = low_reader.drop(low_reader.index[0:5])
third_lowest
```

```python
 print("Movie Which Has Third Lowest Budget : ",third_lowest['original_title'])
```

```python
high_reader = refined_df_budget.nlargest(3, ['budget'])
high_reader
```

```python
third_highest = high_reader.drop(high_reader.index[0:2])
third_highest
```

```python
 print("Movie Which Has Third Highest Budget : ",third_highest['original_title'])
```
